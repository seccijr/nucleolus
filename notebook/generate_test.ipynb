```python
import os
import re
import numpy as np
from subprocess import check_output as qx
```

```python
try:
    file_dir = os.path.dirname(os.path.abspath(__file__))
except NameError:
    file_dir = "."
```

```python
results_directory = os.path.join(file_dir, "..", "test", "results")
executable_directory = os.path.join(file_dir, "..", "executable")
games = []

for file_name in os.listdir(results_directory):
    full_path = os.path.join(results_directory, file_name)
    match = re.match(r"v(\d+)n(\d+)j(\d+)\.txt", file_name)
    if os.path.isfile(full_path) and match:
        file = open(full_path, 'r')
        type = int(match.group(1))
        n = int(match.group(2))
        if (n > 10 or not type in [1, 4, 2]):
            continue
        lines = file.readlines()
        if len(lines) > 0:
            seed = int(lines[0])
            result = np.zeros(n)
            i = len(lines)
            k = n
            found_nucleolus = False
            for j in range(len(lines)):
                i = i - 1
                line = float(lines[i])
                if not found_nucleolus:
                    found_nucleolus = line == 1
                else:
                    k = k - 1
                    result[k] = line
                if k == 0:
                    break
            if found_nucleolus and k == 0:
                cmd = os.path.join(executable_directory, "OUTPUT_GAME.exe")
                output = qx([cmd, str(n), str(type), str(seed)])
                output_lines = map(str, output.decode().splitlines())
                v = ", ".join(output_lines)
                nucleolus = ", ".join(map(str, result))
                games.append(f"{{{n}, \"{file_name}\", {{{v}}}, {{{nucleolus}}}}}")
```

```python
print(",\n".join(games))
```
